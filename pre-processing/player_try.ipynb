```python
import pandas as pd
import os

# Replace this with the path to your downloads folder
downloads_folder = '/Users/poojithabalamurugan/Downloads/'

# Initialize an empty list to hold dataframes
dfs = []

# Iterate over each year and read the corresponding CSV file
for year in range(2017, 2024):
    file_path = os.path.join(downloads_folder, f'{year}_LoL_esports_match_data_from_OraclesElixir.csv')
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all dataframes into one
df = pd.concat(dfs, ignore_index=True)
```

```python
# Filter out only the rows with complete data
df = df[df["datacompleteness"] == "complete"]
```

```python
import pandas as pd

# Define the aggregations for player statistics
aggregations = {
    'gameid': 'count',  # Total number of games
    'result': 'sum',  # Total number of wins (assuming 'result' == 1 means a win)
    'kills': 'sum',
    'deaths': 'sum',
    'assists': 'sum',
    'teamkills': 'sum',  # Required for calculating kill participation
    'teamdeaths': 'sum',  # Required for calculating death percent
    'totalgold': 'sum',  # Total gold for gold percent calculation
    'damagetochampions': 'sum',  # Total damage to champions for DPM calculation
    'wardsplaced': 'sum',  # Total wards placed for WPM calculation
    'wardskilled': 'sum',  # Total wards killed for WCPM calculation
    'total cs': 'sum',  # Total creep score for CS and CSM calculation
    'gamelength': 'sum',  # Total game length in minutes for per minute calculations
    'firstbloodkill': 'sum',  # Sum of first bloods achieved by the player
}

# Aggregate data by 'playername' to calculate career stats
player_stats = df.groupby('playerid').agg(aggregations)
player_stats.columns = ['total_games', 'total_wins', 'total_kills', 'total_deaths', 'total_assists', 'team_kills', 'team_deaths',
                        'total_gold', 'total_damage', 'total_wards_placed', 'total_wards_killed', 'total_creep_score', 'total_game_minutes',
                        'total_first_bloods']

# Calculate additional statistics
player_stats['total_losses'] = player_stats['total_games'] - player_stats['total_wins']
player_stats['kda_ratio'] = (player_stats['total_kills'] + player_stats['total_assists']) / player_stats['total_deaths'].replace(0, 1)
player_stats['kd_ratio'] = player_stats['total_kills'] / player_stats['total_deaths'].replace(0, 1)
player_stats['kill_participation'] = (player_stats['total_kills'] + player_stats['total_assists']) / player_stats['team_kills']
player_stats['death_percent'] = player_stats['total_deaths'] / player_stats['team_deaths']
player_stats['kill_share'] = player_stats['total_kills'] / player_stats['team_kills']
player_stats['creep_score'] = player_stats['total_creep_score']
player_stats['creeps_per_minute'] = player_stats['total_creep_score'] / (player_stats['total_game_minutes'] / 60)
player_stats['gold_percent'] = player_stats['total_gold'] / df['totalgold'].sum()
player_stats['damage_per_minute'] = player_stats['total_damage'] / (player_stats['total_game_minutes'] / 60)
player_stats['first_blood'] = player_stats['total_first_bloods']
player_stats['wards_per_minute'] = player_stats['total_wards_placed'] / (player_stats['total_game_minutes'] / 60)
player_stats['wards_cleared_per_minute'] = player_stats['total_wards_killed'] / (player_stats['total_game_minutes'] / 60)

# Reset index to turn the index into a column
player_stats.reset_index(inplace=True)

# Merge the aggregated player stats back to the original DataFrame on 'playername'
df = df.merge(player_stats, on='playerid', how='left')
```

```python
# Filter out only the rows with complete data
team_records = df[df["datacompleteness"] == "complete"]

# Filter out only the rows for team records (not individual player records)
team_records = df[df["participantid"] > 10]

# Drop unnecessary columns
team_records = team_records[
    [
        "gameid",
        "participantid",
        "pick1",
        "pick2",
        "pick3",
        "pick4",
        "pick5",
        "result",
    ]
]

team_records = team_records.dropna()

# team_records.drop(
#     columns=[
#         "firsttothreetowers", "mountains", "towers", "firsttower", "firstbaron", 
#         "firstdragon", "dragons", "opp_dragons", "elementaldrakes", 
#         "opp_elementaldrakes", "infernals", "clouds", "firstmidtower", "oceans", 
#         "chemtechs", "hextechs", "dragons (type unknown)", "elders", "opp_elders", 
#         "firstherald", "heralds", "opp_heralds", "opp_void_grubs", "opp_towers", 
#         "turretplates"
#     ],
#     errors='ignore',  # Ignore columns that are not found in the DataFrame
#     inplace=True
# )
```

```python
# Add opponent's picks as new columns
for i, row in team_records.iterrows():
    # Find the opponent's row for the same game
    opponent_row = team_records[
        (team_records["gameid"] == row["gameid"])
        & (team_records["participantid"] != row["participantid"])
    ]
    if not opponent_row.empty:  # Check if the opponent row exists
        for pick in ["pick1", "pick2", "pick3", "pick4", "pick5"]:
            team_records.loc[i, f"opp_{pick}"] = opponent_row[pick].values[0]
    else:
        # Handle the case where no opponent row is found
        for pick in ["pick1", "pick2", "pick3", "pick4", "pick5"]:
            team_records.loc[i, f"opp_{pick}"] = None  # or use 'Unknown', etc.
```

```python
# Define the columns for picks and opponent picks
pick_columns = ["pick1", "pick2", "pick3", "pick4", "pick5"]
opp_pick_columns = ["opp_pick1", "opp_pick2", "opp_pick3", "opp_pick4", "opp_pick5"]

# Convert all pick columns in team_records and player_stats to string to ensure consistent data types for merging
for col in pick_columns + opp_pick_columns:
    team_records[col] = team_records[col].astype(str)

# Assuming player_stats index is also a string, if not convert it:
player_stats.index = player_stats.index.astype(str)

# Iterate over each pick column
for pick_col, opp_pick_col in zip(pick_columns, opp_pick_columns):
    # Join the career stats for the team's picks
    team_records = team_records.join(
        player_stats, on=pick_col, rsuffix=f"_stats_{pick_col}"
    )

    # Join the career stats for the opponent's picks
    team_records = team_records.join(
        player_stats, on=opp_pick_col, rsuffix=f"_stats_{opp_pick_col}"
    )
```

```python
# Concatenate all pick columns and opp_pick columns into two separate series
team_picks = pd.concat([team_records[f"pick{i}"] for i in range(1, 6)])
opp_picks = pd.concat([team_records[f"opp_pick{i}"] for i in range(1, 6)])

# One-hot encode the concatenated series
team_picks_encoded = pd.get_dummies(team_picks, prefix="TeamPick")
opp_picks_encoded = pd.get_dummies(opp_picks, prefix="OppPick")

# Group by the index (gameid) and sum the one-hot encoded columns
team_picks_encoded = team_picks_encoded.groupby(team_picks_encoded.index).sum()
opp_picks_encoded = opp_picks_encoded.groupby(opp_picks_encoded.index).sum()

# Join the one-hot encoded columns with the team_records dataframe
final_player_df = team_records.join(team_picks_encoded).join(opp_picks_encoded)
```

```python
# Rename `kills`, `deaths`, and `assists` columns to end_with `_stats_pick1`
final_player_df.rename(
    columns={
        "kills": "kills_stats_pick1",
        "deaths": "deaths_stats_pick1",
        "assists": "assists_stats_pick1",
    },
    inplace=True,
)
```

```python
# Save the final dataframe to a CSV file
final_player_df.to_csv("player.csv", index=False)
```
