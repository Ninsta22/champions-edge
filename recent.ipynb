#### subset the data for the relevant years

```python
# Filter the DataFrame for the years 2019 to 2023
subset_old_df = concatenated_df[concatenated_df['year'].between(2019, 2023)]
# Calculate the number of unique players in the subset
unique_players = subset_old_df['playerid'].nunique()
print("Number of unique players from 2019 to 2023:", unique_players)
```

#### convert to lowercase

```python
# Convert 'Player' column in new dataset and 'playername' column in old dataset to lowercase
new_df['Player'] = new_df['Player'].str.lower()
subset_old_df['playername'] = subset_old_df['playername'].str.lower()
```

```python
# Remove rows where 'Player' contains any numbers
new_df = new_df[~new_df['Player'].str.contains('\d')]
```

```python
# Check what player names are in the new dataset but not in the old dataset
players_not_in_old = new_df[~new_df['Player'].isin(subset_old_df['playername'])]['Player']
print(players_not_in_old)

# Remove these players from the new dataset
new_df = new_df[new_df['Player'].isin(subset_old_df['playername'])]
```

```python
# Merge the datasets on player names
merged_df = subset_old_df.merge(new_df, left_on='playername', right_on='Player', how='left')
merged_df.head()
```

```python
# Replace NaN values with median or mode
for column in new_df.columns:
    # Skip player name to avoid filling with median or mode
    if column not in ['player_name']:
        # If the column is numeric, use median; if categorical, use mode
        if merged_df[column].dtype == 'float64' or merged_df[column].dtype == 'int64':
            merged_df[column].fillna(merged_df[column].median(), inplace=True)
        else:
            merged_df[column].fillna(merged_df[column].mode()[0], inplace=True)  # Mode returns a Series, use [0] to get value
```

#### Aggregate player stats as average values using groupby on 'playerid'

```python

player_stats_df = merged_df.groupby('playerid').agg({
    'kills': 'mean',
    'deaths': 'mean',
    'assists': 'mean',
    'damagetochampions': 'mean',
    'dpm': 'mean',
    'totalgold': 'mean',
    'earnedgold': 'mean',
    'total cs': 'mean',
    'minionkills': 'mean',
    'monsterkills': 'mean',
    'monsterkillsownjungle': 'mean',
    'monsterkillsenemyjungle': 'mean',
    'cspm': 'mean',
    'visionscore': 'mean',
    'vspm': 'mean',
    'wardsplaced': 'mean',
    'wpm': 'mean',
    'wardskilled': 'mean',
    'wcpm': 'mean',
    'controlwardsbought': 'mean',
    'goldspent': 'mean',
    'gspd': 'mean'  # gold spend difference, if relevant
}).reset_index()

# Print to check the new DataFrame structure
print(player_stats_df.head())
```

#### Group by 'gameid' and 'teamid' to calculate the average stats for players in each team

```python

team_stats = merged_df.groupby(['gameid', 'teamid']).agg({
    'kills': 'mean',
    'deaths': 'mean',
    'assists': 'mean',
    'damagetochampions': 'mean',
    'dpm': 'mean',
    'totalgold': 'mean',
    'earnedgold': 'mean',
    'total cs': 'mean',
    'minionkills': 'mean',
    'monsterkills': 'mean',
    'monsterkillsownjungle': 'mean',
    'monsterkillsenemyjungle': 'mean',
    'cspm': 'mean',
    'visionscore': 'mean',
    'vspm': 'mean',
    'wardsplaced': 'mean',
    'wpm': 'mean',
    'wardskilled': 'mean',
    'wcpm': 'mean',
    'controlwardsbought': 'mean',
    'goldspent': 'mean',
    'gspd': 'mean',  # gold spend difference, if relevant
    # Add more team-relevant stats here
}).reset_index()

# Rename columns to reflect that these are team averages
team_stats.rename(columns=lambda x: f'{x}_team_avg' if x not in ['gameid', 'teamid'] else x, inplace=True)

# Now you have team-level statistics that you can merge back to your original dataframe
# to associate each game and team with their corresponding average stats.
# Assuming 'merged_df' has 'gameid' and 'teamid' columns to allow this merge.
merged_with_team_stats = merged_df.merge(team_stats, on=['gameid', 'teamid'], how='left')
```
